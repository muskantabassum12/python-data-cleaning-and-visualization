```python
df = pd.read_excel("/kaggle/input/customer-call-dataset/Customer Call List.xlsx")
df
```

```python
#removes duplicates
df = df.drop_duplicates()
df
```

```python
#columns drop 
#df = df.drop(columns='Not_Useful_Column')
df = df.drop('Not_Useful_Column',axis=1)
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df["Last_Name"]
#"123._/" means that any leading occurrence of the characters 1, 2, 3, ., _, or / will be removed.
#3_abc", "/.123_def", "123./ghi", "_jkl
#0      abc
#1      def
#2      ghi
#3      jkl
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('|','')
df['Phone_Number'] = df['Phone_Number'].str.replace('/','')
df['Phone_Number'] = df['Phone_Number'].str.replace('--','-')
df['Phone_Number'] = df['Phone_Number'].str.replace('nan-','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na-','')
df
```

```python
def make(x):
    x = str(x)
    return x[0:3] + '-' + x[3:6] + '-' + x[6:10]

df['Phone_Number'] = df['Phone_Number'].apply(make)
df 
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('|','')
df['Phone_Number'] = df['Phone_Number'].str.replace('/','')
df['Phone_Number'] = df['Phone_Number'].str.replace('--','-')
df['Phone_Number'] = df['Phone_Number'].str.replace('nan-','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na-','')
df
```

```python
#splitting a column data based on comma
df[["Street","State","Zipcode"]] = df["Address"].str.split(",", n=2, expand=True)
df
```

```python
df = df.drop('Address', axis=1)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)

df
```
